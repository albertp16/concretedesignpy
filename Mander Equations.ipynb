```python
##Using Fig. 6.4 from Mander et. al (1988), we can determine ratioo of f'cc/f'c which in this case is 1.6 

def ecc(eco,fpcc,fc):
    A = ((fpcc/fc) - 1) 
    B = 1 + 5*A
    results = eco*B
    return results
                   
e_cc = ecc(0.002,48,30)
```

```python
def Esec(fpcc,ecc):
    return fpcc/ecc
def Ec(fc):
    return 5000*math.sqrt(fc)
def nValue(Ec,Esec):
    A = Ec/(Ec - Esec)
    B = 1 + (5*A)
    return A
def stressIndex(n,ec,ecc,fpcc):
    A = ec/ecc
    B = pow(A,n)
    C = n*A
    D = (n-1)+B
    E = C/D
    F = fpcc*E
    return F 


strain_arr=[]
stress_arr=[]
i = 0 
while i <= e_c:
    fc_index = stressIndex(1.28,i,0.008,48)
    strain_arr.append(i)
    stress_arr.append(fc_index)
    i = i + 0.001

fig, ax = plt.subplots()
ax.plot(strain_arr, stress_arr)

ax.set(xlabel='Strain', ylabel='Stress',
       title='Manders')
ax.grid()

fig.savefig("test.png")
plt.show()
```
