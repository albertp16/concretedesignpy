```python
##Reading excel reference 
# df = pd.read_excel('ref.xlsx')
# df

midas_nodes = set(df['Node'])
midas_case = set(df['Load'])
# print(midas_case)

data_length = len(df)
total_fdn = len(midas_nodes)

data = []

dl_arr = []
sdl_arr = []
ll_arr = []
ex_arr = []
exp_arr = []
exn_arr = []
ey_arr = []
eyp_arr = []
eyn_arr = []
ro_arr = []
rm_arr = []

for i in midas_case: 
    j = 0
    while j < data_length:
        if i == df['Load'][j]:
            if(i == 'RO(RS)'): ro_arr.append([df['FZ (kN)'][j]*-1,df['MX (kN*m)'][j]*-1,df['MY (kN*m)'][j]*-1])
            if(i == 'Ey'):ey_arr.append([df['FZ (kN)'][j]*-1,df['MX (kN*m)'][j]*-1,df['MY (kN*m)'][j]*-1])
            if(i == 'Ex'):ex_arr.append([df['FZ (kN)'][j]*-1,df['MX (kN*m)'][j]*-1,df['MY (kN*m)'][j]*-1])
            if(i == 'Ey +'):exp_arr.append([df['FZ (kN)'][j]*-1,df['MX (kN*m)'][j]*-1,df['MY (kN*m)'][j]*-1])
            if(i == 'DL'):dl_arr.append([df['FZ (kN)'][j]*-1,df['MX (kN*m)'][j]*-1,df['MY (kN*m)'][j]*-1])
            if(i == 'SDL'):sdl_arr.append([df['FZ (kN)'][j]*-1,df['MX (kN*m)'][j]*-1,df['MY (kN*m)'][j]*-1])
            if(i == 'Ex -'):exn_arr.append([df['FZ (kN)'][j]*-1,df['MX (kN*m)'][j]*-1,df['MY (kN*m)'][j]*-1])
            if(i == 'LL'):ll_arr.append([df['FZ (kN)'][j]*-1,df['MX (kN*m)'][j]*-1,df['MY (kN*m)'][j]*-1])
            if(i == 'RM(RS)'):rm_arr.append([df['FZ (kN)'][j]*-1,df['MX (kN*m)'][j]*-1,df['MY (kN*m)'][j]*-1])
            if(i == 'Ey -'):eyn_arr.append([df['FZ (kN)'][j]*-1,df['MX (kN*m)'][j]*-1,df['MY (kN*m)'][j]*-1])
            if(i == 'Ex +'):eyp_arr.append([df['FZ (kN)'][j]*-1,df['MX (kN*m)'][j]*-1,df['MY (kN*m)'][j]*-1])
#             if(i == 'Lr'):ro_arr.append([df['FZ (kN)'][j],df['MX (kN*m)'][j],df['MY (kN*m)'][j]])
        j += 1


print(ro_arr)
```

```python
input_string = '''
LOADCASES
1
Dead DL
Dead SDL
Live LL
EQ Ex
EQ "Ex +"
EQ "Ex -"
EQ Ey
EQ "Ey +"
EQ "Ey -"
EQ RM
EQ RO

POINTS
'''
no_nodes = len(nodes)
# no_nodes
i = 0;
while i < no_nodes:
    id_index = nodes[i][0]
    x_index = nodes[i][1]
    y_index = nodes[i][2]
    ## nodal output
    nodal_data = f'{id_index} {x_index} {y_index} \n'
    input_string += nodal_data
    i += 1

input_string += ' \n'
input_string += 'LOADS'    

i = 0;
while i < no_nodes:
    id_index = nodes[i][0]
#     x_index = nodes[i][1]
#     y_index = nodes[i][2]
    ## nodal output
    nodal_data = f'''
{id_index} DL {dl_arr[i][0]} {dl_arr[i][1]} {dl_arr[i][2]}
{id_index} SDL {sdl_arr[i][0]} {sdl_arr[i][1]} {sdl_arr[i][2]}
{id_index} LL {ll_arr[i][0]} {ll_arr[i][1]} {ll_arr[i][2]}
{id_index} Ex  {ex_arr[i][0]} {ex_arr[i][1]} {ex_arr[i][2]}
{id_index} "Ex +" {exp_arr[i][0]} {exp_arr[i][1]} {exp_arr[i][2]}
{id_index} "Ex -" {exn_arr[i][0]} {exn_arr[i][1]} {exn_arr[i][2]}
{id_index} Ey {ey_arr[i][0]} {ey_arr[i][1]} {ey_arr[i][2]}
{id_index} "Ey +" {eyp_arr[i][0]} {eyp_arr[i][1]} {eyp_arr[i][2]}
{id_index} "Ey -" {eyn_arr[i][0]} {eyn_arr[i][1]} {eyn_arr[i][2]}
{id_index} RM {rm_arr[i][0]} {rm_arr[i][1]} {rm_arr[i][2]}
{id_index} RO {ro_arr[i][0]} {ro_arr[i][1]} {ro_arr[i][2]}'''
    input_string += nodal_data
    i += 1

print(input_string)
```
